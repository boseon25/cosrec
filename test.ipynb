```python
# -*- coding: utf-8 -*-
import pandas as pd
import sqlite3
import re

# 1. CSV ë¶ˆëŸ¬ì˜¤ê¸°
input_path = r"C:\Users\user\Desktop\Github\íš¨ê³¼DB_ì„±ë¶„ì„¤ëª…_gemini.csv"
df = pd.read_csv(input_path)

# 2. ì„¤ëª… ì •ë¦¬ í•¨ìˆ˜ ì •ì˜
def clean_description(ingredient, text):
    if pd.isna(text):
        return ""
    
    # 1) \n, \\n â†’ ì¤„ë°”ê¿ˆ ë˜ëŠ” ê³µë°±ìœ¼ë¡œ ì •ë¦¬
    text = text.replace("\\n", "\n").replace("\n", " ").strip()

    # 2) ì„±ë¶„ëª…, ##, ì„±ë¶„ ë¶„ì„ ë“± ì œê±°
    patterns_to_remove = [
        r"#*\s*" + re.escape(ingredient) + r"\s*ì„±ë¶„ ë¶„ì„",
        r"#*\s*" + re.escape(ingredient),
        r"#*\s*ì„±ë¶„ ë¶„ì„"
    ]
    for pattern in patterns_to_remove:
        text = re.sub(pattern, "", text, flags=re.IGNORECASE)

    # 3) ê³µë°± ì •ë¦¬
    text = re.sub(r"\s+", " ", text).strip()

    return text

# 3. ì„¤ëª… ì—´ ì •ë¦¬
df['ì„¤ëª…'] = df.apply(lambda row: clean_description(row['ì„±ë¶„ëª…'], row['ì„¤ëª…']), axis=1)

# 4. ê²°ê³¼ CSV ì €ì¥
output_csv_path = r"C:\Users\user\Desktop\Github\íš¨ê³¼DB_ì„±ë¶„ì„¤ëª…_ì •ë¦¬ë¨.csv"
df.to_csv(output_csv_path, index=False, encoding="utf-8-sig", lineterminator='\n')

# 5. SQLite ì €ì¥
conn = sqlite3.connect("skincare_ingredient_cleaned.db")
df.to_sql("ingredient_descriptions", conn, if_exists="replace", index=False)
conn.close()

print("âœ… ì„¤ëª… ì •ë¦¬ ì™„ë£Œ ë° CSV/DB ì €ì¥ ì™„ë£Œ")
```

```python
import pandas as pd

# CSV íŒŒì¼ ê²½ë¡œë“¤
ingredient_paths = [
    r"C:\Users\user\Desktop\Github\cosrec\product_csv\ingredient\Ingredients_list.csv",
    r"C:\Users\user\Desktop\Github\cosrec\product_csv\ingredient\Ingredients_list1.csv",
    r"C:\Users\user\Desktop\Github\cosrec\product_csv\ingredient\Ingredients_list2.csv",
    r"C:\Users\user\Desktop\Github\cosrec\product_csv\ingredient\Ingredients_list3.csv"
]

merged_rows = []

for path in ingredient_paths:
    with open(path, "r", encoding="utf-8") as f:
        for line in f:
            # 1. ì¤„ ë ê°œí–‰ ì œê±° ë° íŠ¹ìˆ˜ ë¬¸ì í†µì¼
            line = line.strip().replace("@", ",")
            # 2. ì‰¼í‘œ ê¸°ì¤€ split
            parts = [part.strip() for part in line.split(",") if part.strip() != ""]
            if len(parts) < 2:
                continue  # ì„±ë¶„ ì—†ëŠ” ê²½ìš° skip
            product = parts[0]
            ingredients = parts[1:]
            merged_rows.append([product] + ingredients)

# ë³‘í•©ëœ ë°ì´í„°ë¥¼ DataFrameìœ¼ë¡œ ë³€í™˜
max_len = max(len(row) for row in merged_rows)
columns = ['ì œí’ˆëª…'] + [f'ì„±ë¶„{i+1}' for i in range(max_len - 1)]
merged_ingredients_df = pd.DataFrame(merged_rows, columns=columns)

# ê²°ê³¼ ì €ì¥ (ì¤„ë°”ê¿ˆ í†µì¼ í¬í•¨)
merged_ingredients_df.to_csv("Merged_Ingredients_List.csv", index=False, lineterminator='\n')

print("âœ… ë³‘í•© ì™„ë£Œ:", merged_ingredients_df.shape)
```

```python
import pandas as pd
import re

# 1. CSV ë¶ˆëŸ¬ì˜¤ê¸°
df = pd.read_csv("Merged_Ingredients_List.csv")

# 2. ì •ì œëœ ì„±ë¶„ ë¦¬ìŠ¤íŠ¸ ì €ì¥ìš© ì§‘í•©
ingredient_set = set()

# 3. ì •ê·œí‘œí˜„ì‹ íŒ¨í„´
remove_prefix_pattern = re.compile(r'^[\d\s\.\-():#a-zA-Zê°€-í£]*[:)\-]?\s*')  # ì ‘ë‘ì–´ ì œê±°
unit_pattern = re.compile(r'\b\d+(\.\d+)?\s*(ppm|ppb|%|IU|mg|g|ã|ã|Î¼g)\b', re.IGNORECASE)  # ë†ë„ ì œê±°
only_unit_pattern = re.compile(r'^\d+(ppm|ppb|%|IU|mg|g|ã|ã|Î¼g)$', re.IGNORECASE)

# 4. ì „ì²´ ì—´ì—ì„œ ì„±ë¶„ ì¶”ì¶œ
for col in df.columns[1:]:
    for cell in df[col].dropna().astype(str):
        # êµ¬ë¶„ì ë¶„í•  (ì‰¼í‘œ, â€¢, /, [], # ë“± ë‹¤ì–‘í•œ ê²½ìš°)
        items = re.split(r"[â€¢,/\[\]\n#]", cell)
        for item in items:
            cleaned = item.strip()
            if not cleaned:
                continue
            # ë¶ˆí•„ìš”í•œ ì ‘ë‘ì–´ ì œê±°
            cleaned = remove_prefix_pattern.sub('', cleaned)
            # ê´„í˜¸ ì•ˆ ë‹¨ìœ„ ì œê±°
            cleaned = re.sub(r'\((.*?)\)', '', cleaned)
            # ìˆœìˆ˜ ë‹¨ìœ„ë§Œ ë‚¨ì€ ê²½ìš° ì œê±°
            if only_unit_pattern.match(cleaned):
                continue
            # ë„ˆë¬´ ì§§ê±°ë‚˜ íŠ¹ìˆ˜ë¬¸ìë§Œ ìˆëŠ” ê²½ìš° ì œê±°
            if len(cleaned) >= 2 and not re.match(r'^[\d\W_]+$', cleaned):
                ingredient_set.add(cleaned)

# 5. ì¤‘ë³µ ì œê±° ë° ì •ë ¬
unique_ingredients = sorted(ingredient_set)

# 6. ì„±ë¶„ ì‚¬ì „ í…œí”Œë¦¿ ìƒì„±
columns = [
    "ì„±ë¶„ëª…", "íš¨ê³¼", "ë¶€ì‘ìš©", "ì£¼ì˜ì‚¬í•­", "ì í•©í•œ í”¼ë¶€ íƒ€ì…",
    "ê¶Œì¥ ë†ë„", "í•¨ê»˜ ì‚¬ìš©í•˜ë©´ ì•ˆ ë˜ëŠ” ì„±ë¶„", "í¬í•¨ëœ ëŒ€í‘œ ì œí’ˆ", "ì‚¬ìš© ì‹œê¸°/ê³„ì ˆ"
]
template_df = pd.DataFrame(columns=columns)
template_df["ì„±ë¶„ëª…"] = unique_ingredients

# 7. ì €ì¥
template_df.to_csv("Ingredient_Encyclopedia_Template.csv", index=False, encoding='utf-8-sig')
print(f"âœ… ì „ì²˜ë¦¬ í›„ ì„±ë¶„ ìˆ˜: {len(template_df)}ê°œ")
```
