## 깔끔히 정리 코드

```python
# -*- coding: utf-8 -*-
import pandas as pd
import sqlite3
import re

# 1. CSV 불러오기
input_path = r"C:\Users\user\Desktop\Github\효과DB_성분설명_gemini.csv"
df = pd.read_csv(input_path)

# 2. 설명 정리 함수 정의
def clean_description(ingredient, text):
    if pd.isna(text):
        return ""
    
    # 1) \n, \\n → 줄바꿈 또는 공백으로 정리
    text = text.replace("\\n", "\n").replace("\n", " ").strip()

    # 2) 성분명, ##, 성분 분석 등 제거
    patterns_to_remove = [
        r"#*\s*" + re.escape(ingredient) + r"\s*성분 분석",
        r"#*\s*" + re.escape(ingredient),
        r"#*\s*성분 분석"
    ]
    for pattern in patterns_to_remove:
        text = re.sub(pattern, "", text, flags=re.IGNORECASE)

    # 3) 공백 정리
    text = re.sub(r"\s+", " ", text).strip()

    return text

# 3. 설명 열 정리
df['설명'] = df.apply(lambda row: clean_description(row['성분명'], row['설명']), axis=1)

# 4. 결과 CSV 저장
output_csv_path = r"C:\Users\user\Desktop\Github\효과DB_성분설명_정리됨.csv"
df.to_csv(output_csv_path, index=False, encoding="utf-8-sig", lineterminator='\n')

# 5. SQLite 저장
conn = sqlite3.connect("skincare_ingredient_cleaned.db")
df.to_sql("ingredient_descriptions", conn, if_exists="replace", index=False)
conn.close()

print("✅ 설명 정리 완료 및 CSV/DB 저장 완료")
```

```python
import pandas as pd

# CSV 파일 경로들
ingredient_paths = [
    r"C:\Users\user\Desktop\Github\cosrec\product_csv\ingredient\Ingredients_list.csv",
    r"C:\Users\user\Desktop\Github\cosrec\product_csv\ingredient\Ingredients_list1.csv",
    r"C:\Users\user\Desktop\Github\cosrec\product_csv\ingredient\Ingredients_list2.csv",
    r"C:\Users\user\Desktop\Github\cosrec\product_csv\ingredient\Ingredients_list3.csv"
]

merged_rows = []

for path in ingredient_paths:
    with open(path, "r", encoding="utf-8") as f:
        for line in f:
            # 1. 줄 끝 개행 제거 및 특수 문자 통일
            line = line.strip().replace("@", ",")
            # 2. 쉼표 기준 split
            parts = [part.strip() for part in line.split(",") if part.strip() != ""]
            if len(parts) < 2:
                continue  # 성분 없는 경우 skip
            product = parts[0]
            ingredients = parts[1:]
            merged_rows.append([product] + ingredients)

# 병합된 데이터를 DataFrame으로 변환
max_len = max(len(row) for row in merged_rows)
columns = ['제품명'] + [f'성분{i+1}' for i in range(max_len - 1)]
merged_ingredients_df = pd.DataFrame(merged_rows, columns=columns)

# 결과 저장 (줄바꿈 통일 포함)
merged_ingredients_df.to_csv("Merged_Ingredients_List.csv", index=False, lineterminator='\n')

print("✅ 병합 완료:", merged_ingredients_df.shape)
```

```python
import pandas as pd
import re

# 1. CSV 불러오기
df = pd.read_csv("Merged_Ingredients_List.csv")

# 2. 정제된 성분 리스트 저장용 집합
ingredient_set = set()

# 3. 정규표현식 패턴
remove_prefix_pattern = re.compile(r'^[\d\s\.\-():#a-zA-Z가-힣]*[:)\-]?\s*')  # 접두어 제거
unit_pattern = re.compile(r'\b\d+(\.\d+)?\s*(ppm|ppb|%|IU|mg|g|㎍|㎎|μg)\b', re.IGNORECASE)  # 농도 제거
only_unit_pattern = re.compile(r'^\d+(ppm|ppb|%|IU|mg|g|㎍|㎎|μg)$', re.IGNORECASE)

# 4. 전체 열에서 성분 추출
for col in df.columns[1:]:
    for cell in df[col].dropna().astype(str):
        # 구분자 분할 (쉼표, •, /, [], # 등 다양한 경우)
        items = re.split(r"[•,/\[\]\n#]", cell)
        for item in items:
            cleaned = item.strip()
            if not cleaned:
                continue
            # 불필요한 접두어 제거
            cleaned = remove_prefix_pattern.sub('', cleaned)
            # 괄호 안 단위 제거
            cleaned = re.sub(r'\((.*?)\)', '', cleaned)
            # 순수 단위만 남은 경우 제거
            if only_unit_pattern.match(cleaned):
                continue
            # 너무 짧거나 특수문자만 있는 경우 제거
            if len(cleaned) >= 2 and not re.match(r'^[\d\W_]+$', cleaned):
                ingredient_set.add(cleaned)

# 5. 중복 제거 및 정렬
unique_ingredients = sorted(ingredient_set)

# 6. 성분 사전 템플릿 생성
columns = [
    "성분명", "효과", "부작용", "주의사항", "적합한 피부 타입",
    "권장 농도", "함께 사용하면 안 되는 성분", "포함된 대표 제품", "사용 시기/계절"
]
template_df = pd.DataFrame(columns=columns)
template_df["성분명"] = unique_ingredients

# 7. 저장
template_df.to_csv("Ingredient_Encyclopedia_Template.csv", index=False, encoding='utf-8-sig')
print(f"✅ 전처리 후 성분 수: {len(template_df)}개")
```
